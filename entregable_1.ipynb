<!-- Renombraremos columnas a espaÃ±ol para lograr analizar de manera mas rÃ¡pida respaldando el documento original, tambien ordenaremos las fechas de la muestra de menor a mayor -->

```python
data_esp = data.rename(columns={
    'Date':'Fecha',
    'Location': 'Ubicacion',
    'MinTemp': 'MinTemp',
    'MaxTemp': 'MaxTemp',
    'Rainfall': 'Lluvia',
    'Evaporation': 'Evaporacion',
    'Sunshine': 'Sol',
    'WindGustDir': 'DirRafaga',
    'WindGustSpeed': 'VelRafaga',
    'WindDir9am': 'Dir9am',
    'WindDir3pm': 'Dir3pm',
    'WindSpeed9am': 'Vel9am',
    'WindSpeed3pm': 'Vel3pm',
    'Humidity9am': 'Hum9am',
    'Humidity3pm': 'Hum3pm',
    'Pressure9am': 'Pres9am',
    'Pressure3pm': 'Pres3pm',
    'Cloud9am': 'Nub9am',
    'Cloud3pm': 'Nub3pm',
    'Temp9am': 'Temp9am',
    'Temp3pm': 'Temp3pm',
    'RainToday': 'LluviaHoy',
    'RISK_MM': 'RiesgoMM',
    'RainTomorrow': 'LluviaMan'
})
data_esp['Fecha'] = pd.to_datetime(data_esp['Fecha'])
data_esp = data_esp.sort_values(by='Fecha')
data_esp = data_esp.reset_index(drop=True)
data_esp
```

<!-- Tratamiento de datos nulos -->

```python
data_esp.to_excel("Nulos.xlsx", index=False)
```

```python
print("Valores nulos por columna:")
for feature in data_esp.columns:
    total_nulos = data_esp[feature].isna().sum()
    porcentaje_nulos = (total_nulos / len(data_esp)) * 100
    print(f'{feature}: {total_nulos} valores nulos ({porcentaje_nulos:.2f}%)')
```

```python
# ExploraciÃ³n de valores
print("EstadÃ­sticas de la variable Ubicacion:")
data_esp['Ubicacion'].describe()
```

```python
# EstadÃ­sticas de temperaturas para conocer el clima de Australia
columnas_temp = ['MinTemp', 'MaxTemp']
for col in columnas_temp:
    print(f'\nAnÃ¡lisis de {col}:')
    print(f'MÃ­nimo: {data_esp[col].min()}')
    print(f'MÃ¡ximo: {data_esp[col].max()}')
    print(f'Promedio: {data_esp[col].mean()}')
    print(f'STD: {data_esp[col].std()}')
    print(data_esp[col].quantile([.25, .5, .75]))
```

```python
#Queremos saber donde llueve mÃ¡s
print("10 ubicaciones con mÃ¡s lluvia")
data_esp.groupby('Ubicacion')['Lluvia'].sum().sort_values(ascending=False).head(10)
```

```python
#Queremos saber donde llueve menos
print("10 ubicaciones con menos lluvia")
data_esp.groupby('Ubicacion')['Lluvia'].sum().sort_values(ascending=True).head(10)
```

```python
# Queremos saber la extension de la muestra
data_esp['Fecha'] = pd.to_datetime(data_esp['Fecha'])
fecha_min = data_esp['Fecha'].min().strftime('%d/%m/%Y')
fecha_max = data_esp['Fecha'].max().strftime('%d/%m/%Y')

print(f"Rango de fechas: {fecha_min} a {fecha_max}")
```

```python
# Cuantos aÃ±os?
print("AÃ±os en la data:", data_esp['Fecha'].dt.year.nunique())
```

```python
# Generaremos un histograma de lluvia por aÃ±o para detectar alguna variacion en las precipitaciones
data_esp['Fecha'] = pd.to_datetime(data_esp['Fecha'])
lluvia_por_anio = data_esp.groupby(data_esp['Fecha'].dt.year)['Lluvia'].sum()

plt.Figure(figsize=(10,5))
plt.bar(lluvia_por_anio.index, lluvia_por_anio.values, color = 'skyblue')

plt.xlabel('Anio')
plt.ylabel('Total de lluvia (mm)')
plt.title('Histograma de lluvia por aÃ±o')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.9)

plt.show()
```
