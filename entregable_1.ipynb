<!-- Renombraremos columnas a español para lograr analizar de manera mas rápida respaldando el documento original, tambien ordenaremos las fechas de la muestra de menor a mayor -->

```python
data_esp = data.rename(columns={
    'Date':'Fecha',
    'Location': 'Ubicacion',
    'MinTemp': 'MinTemp',
    'MaxTemp': 'MaxTemp',
    'Rainfall': 'Lluvia',
    'Evaporation': 'Evaporacion',
    'Sunshine': 'Sol',
    'WindGustDir': 'DirRafaga',
    'WindGustSpeed': 'VelRafaga',
    'WindDir9am': 'Dir9am',
    'WindDir3pm': 'Dir3pm',
    'WindSpeed9am': 'Vel9am',
    'WindSpeed3pm': 'Vel3pm',
    'Humidity9am': 'Hum9am',
    'Humidity3pm': 'Hum3pm',
    'Pressure9am': 'Pres9am',
    'Pressure3pm': 'Pres3pm',
    'Cloud9am': 'Nub9am',
    'Cloud3pm': 'Nub3pm',
    'Temp9am': 'Temp9am',
    'Temp3pm': 'Temp3pm',
    'RainToday': 'LluviaHoy',
    'RISK_MM': 'RiesgoMM',
    'RainTomorrow': 'LluviaMan'
})
data_esp['Fecha'] = pd.to_datetime(data_esp['Fecha'])
data_esp = data_esp.sort_values(by='Fecha')
data_esp = data_esp.reset_index(drop=True)
data_esp
```

```python
print("Valores nulos por columna:")
for feature in data_esp.columns:
    total_nulos = data_esp[feature].isna().sum()
    porcentaje_nulos = (total_nulos / len(data_esp)) * 100
    print(f'{feature}: {total_nulos} valores nulos ({porcentaje_nulos:.2f}%)')
```

```python
# Las columnas de temperaturas no las imputaremos, ya que representan un porcentaje menor al 3% de datos nulos
# Imputaremos las columnas numéricas con la mediana
columnas_mediana = ['Pres9am', 'Pres3pm', 'Hum3pm', 'Nub9am', 'Nub3pm', 'VelRafaga']

for col in columnas_mediana:
    if (data_esp[col].isna().sum() / len(data_esp)) * 100 >= 3:
        data_esp[col] = data_esp[col].fillna(data_esp[col].median())
```

```python
# Imputar columnas categóricas con moda
columnas_moda = ['DirRafaga', 'Dir9am', 'Dir3pm']
for col in columnas_moda:
    if (data_esp[col].isna().sum() / len(data_esp)) * 100 > 3:
        data_esp[col] = data_esp[col].fillna(data_esp[col].mode()[0])
```

```python
# Eliminar las columnas Sol y Evaporacion debido a la gran cantidad de valores nulos
data_esp.drop(columns=['Sol', 'Evaporacion','Fecha'], inplace=True)
```

```python
print("Valores nulos por columna:")
for feature in data_esp.columns:
    total_nulos = data_esp[feature].isna().sum()
    porcentaje_nulos = (total_nulos / len(data_esp)) * 100
    print(f'{feature}: {total_nulos} valores nulos ({porcentaje_nulos:.2f}%)')
```

```python
#Necesitamos convertir las variables categoricas a numeros, es por ello que usaremos el metodo Label Encoding
# para convertir categorías en enteros
from sklearn.preprocessing import LabelEncoder
# Para escalar variables numericas
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
#Creamos copia de respaldo
data_encode_esp = data_esp.copy()
cols_categoricas = ['Ubicacion', 'DirRafaga', 'Dir9am', 'Dir3pm','LluviaHoy','LluviaMan']

le = LabelEncoder()
for col in cols_categoricas:
    data_encode_esp[col] = le.fit_transform(data_encode_esp[col].astype(str))

    print(data_encode_esp[cols_categoricas].head())
```

```python
# Verificamos la transformación
data_encode_esp
```

```python
#Obtener la matriz de correlación
correlation_matrix = data_encode_esp.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5,annot_kws={"size": 8},cbar_kws={"shrink": 0.8})
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.yticks(fontsize=9)
plt.title('Matriz de Correlación', fontsize=14, pad=20)
plt.show()
```

```python
#Procederemos a escalar nuestro dataframe, luego de escalar dejamos las columnas con sus nombres y mostramos el resultado

sc = StandardScaler()
scaled_data = sc.fit_transform(data_encode_esp)

df_scaled = pd.DataFrame(scaled_data, columns=data_encode_esp.columns)
df_final = pd.DataFrame(df_scaled)
df_final
```
