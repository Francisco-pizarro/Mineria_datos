<!-- Renombraremos columnas a espaÃ±ol para lograr analizar de manera mas rÃ¡pida respaldando el documento original, tambien ordenaremos las fechas de la muestra de menor a mayor -->

```python
data_esp = data.rename(columns={
    'Date':'Fecha',
    'Location': 'Ubicacion',
    'MinTemp': 'MinTemp',
    'MaxTemp': 'MaxTemp',
    'Rainfall': 'Lluvia',
    'Evaporation': 'Evaporacion',
    'Sunshine': 'Sol',
    'WindGustDir': 'DirRafaga',
    'WindGustSpeed': 'VelRafaga',
    'WindDir9am': 'Dir9am',
    'WindDir3pm': 'Dir3pm',
    'WindSpeed9am': 'Vel9am',
    'WindSpeed3pm': 'Vel3pm',
    'Humidity9am': 'Hum9am',
    'Humidity3pm': 'Hum3pm',
    'Pressure9am': 'Pres9am',
    'Pressure3pm': 'Pres3pm',
    'Cloud9am': 'Nub9am',
    'Cloud3pm': 'Nub3pm',
    'Temp9am': 'Temp9am',
    'Temp3pm': 'Temp3pm',
    'RainToday': 'LluviaHoy',
    'RISK_MM': 'RiesgoMM',
    'RainTomorrow': 'LluviaMan'
})
data_esp['Fecha'] = pd.to_datetime(data_esp['Fecha'])
data_esp = data_esp.sort_values(by='Fecha')
data_esp = data_esp.reset_index(drop=True)
data_esp
```
